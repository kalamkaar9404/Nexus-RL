```python
import yfinance as yf
import pandas as pd
import numpy as np

NIFTY_100_TICKERS = [
    'RELIANCE.NS', 'TCS.NS', 'HDFCBANK.NS', 'INFY.NS',
    'ICICIBANK.NS', 'HINDUNILVR.NS', 'SBIN.NS', 'BAJFINANCE.NS',
    'KOTAKBANK.NS', 'BHARTIARTL.NS'

]

# Time range
START_DATE = '2010-01-01'
END_DATE = '2024-01-01'

daily_data = yf.download(NIFTY_100_TICKERS, start=START_DATE, end=END_DATE, group_by='ticker')


all_stocks_df = []
for ticker in NIFTY_100_TICKERS:
    try:
        df = daily_data[ticker].copy()
        if df.empty:
            continue


        df_monthly = df.resample('ME').last()


        df_monthly['PR'] = df_monthly['Close'].pct_change()

        df_monthly['TV'] = df['Volume'].resample('ME').sum()

        df_monthly['VOL'] = df['Close'].resample('ME').std()

        df_monthly['ticker'] = ticker

        df_monthly['next_return'] = df_monthly['PR'].shift(-1)

        all_stocks_df.append(df_monthly)

    except Exception as e:
        print(f"Could not process {ticker}: {e}")

#DataFrame
full_df = pd.concat(all_stocks_df)


full_df = full_df[['ticker', 'PR', 'TV', 'VOL', 'next_return']].copy()
full_df.index.name = 'date'
full_df.reset_index(inplace=True)

# Drop rows with NaN
full_df.dropna(inplace=True)

# Z-Score Standardization
features = ['PR', 'TV', 'VOL']
for col in features:
    full_df[col] = (full_df[col] - full_df[col].mean()) / full_df[col].std()

# Save the final processed data
full_df.to_csv('india_nifty100.csv', index=False)
print("Saved processed data to india_nifty100.csv")
```

```python
# env/india_env.py
import gym
import numpy as np
import pandas as pd
from collections import deque

class IndiaStockEnv(gym.Env):
    """
    A Gym environment for the AlphaStock model on Indian data.

    It follows the paper's RL optimization:
    - An episode is a fixed "investment" period (e.g., 12 months)[cite: 329].
    - The agent takes an action (selects a portfolio) at each step.
    - The reward is the Sharpe Ratio, given at the *end* of the episode [cite: 149-150, 271, 285].
    """
    def __init__(self, data_path, look_back_k, holding_periods_t, feature_dim,
                 stock_pool, train_test_split_date):

        self.data = pd.read_csv(data_path, parse_dates=['date'])
        self.look_back_k = look_back_k
        self.holding_periods_t = holding_periods_t
        self.feature_dim = feature_dim
        self.stock_pool = stock_pool
        self.num_stocks = len(self.stock_pool)

        # New Feature set
        self.feature_cols = ['PR', 'TV', 'VOL']

        # Split data
        self.train_data = self.data[self.data['date'] < pd.to_datetime(train_test_split_date)]
        self.test_data = self.data[self.data['date'] >= pd.to_datetime(train_test_split_date)]

        self.current_data = self.train_data
        self.unique_dates = sorted(self.current_data['date'].unique())

        #  episode = K (history) + T (investment)
        self.episode_length = self.look_back_k + self.holding_periods_t


        self.start_indices = [
            i for i in range(len(self.unique_dates) - self.episode_length + 1)
        ]

        self.action_space = gym.spaces.Box(low=0, high=1, shape=(self.num_stocks,), dtype=np.float32)
        self.observation_space = gym.spaces.Box(
            low=-np.inf, high=np.inf,
            shape=(self.num_stocks, self.look_back_k, self.feature_dim),
            dtype=np.float32
        )

    def _get_state(self, date_index):
        """
        Get the state: (num_stocks, K, num_features)
        This is the history of K holding periods before date_index.
        """
        state = np.zeros(self.observation_space.shape, dtype=np.float32)

        # Get the K previous dates
        history_dates = self.unique_dates[date_index - self.look_back_k : date_index]

        # Filter data for these dates
        state_data = self.current_data[self.current_data['date'].isin(history_dates)]

        for i, ticker in enumerate(self.stock_pool):
            stock_data = state_data[state_data['ticker'] == ticker].sort_values('date')
            if len(stock_data) == self.look_back_k:
                state[i, :, :] = stock_data[self.feature_cols].values
            # else: leave as zeros (handles missing data)

        return state

    def reset(self):
        self.current_step = 0

        self.episode_start_index = np.random.choice(self.start_indices)

        self.current_date_index = self.episode_start_index + self.look_back_k

        self.period_returns = []

        return self._get_state(self.current_date_index)

    def step(self, action):

        current_date = self.unique_dates[self.current_date_index]

        returns_data = self.current_data[self.current_data['date'] == current_date]

        #  returns vector z_t
        z_t = np.zeros(self.num_stocks, dtype=np.float32)
        for i, ticker in enumerate(self.stock_pool):
            val = returns_data[returns_data['ticker'] == ticker]['next_return'].values
            if len(val) > 0:
                z_t[i] = val[0]

        #portfolio return
        R_t = np.dot(action, z_t)

        self.period_returns.append(R_t)


        self.current_step += 1
        self.current_date_index += 1

        done = False
        reward = 0.0

        if self.current_step == self.holding_periods_t:
            # Episode
            done = True

            # Sharpe Ratio (H_T)
            if len(self.period_returns) > 1 and np.std(self.period_returns) != 0:
                A_T = np.mean(self.period_returns)
                V_T = np.std(self.period_returns)
                reward = A_T / V_T
            else:
                reward = 0.0

        next_state = self._get_state(self.current_date_index) if not done else None

        return next_state, reward, done, {}
```

```python
# model.py
import torch
import torch.nn as nn
import torch.nn.functional as F

#Hyperparameters
LSTM_H_DIM = 128
ATTN_W_DIM = 64
LSTM_DROPOUT = 0.2
LSTM_LAYER = 1
CANN_HEADS = 4

#feature cnt
FEATURE_DIM = 3
class StdAttn(nn.Module):
    def __init__(self, input_size, att_dim, dropout):
        super().__init__()
        self.attention_dim = att_dim
        self.query_transform = nn.Linear(input_size, att_dim)
        self.key_transform = nn.Linear(input_size, att_dim)
        self.attn_score = nn.Sequential(
            nn.Tanh(),
            nn.Dropout(p=dropout),
            nn.Linear(att_dim, 1),
            nn.Softmax(dim=-1)
        )

    def forward(self, q, k, v):
        embed_dim = q.shape[-1]
        bsz = q.shape[0]
        num_queries = q.shape[-2]
        if not k.shape[-1] == embed_dim or not v.shape[-1] == embed_dim:
            raise ValueError("Assume identical embedding dim for q, k, v")
        q = q.reshape(num_queries, bsz, embed_dim)
        trans_q = self.query_transform(q)
        trans_k = self.key_transform(k)
        attn_input = trans_k + trans_q
        attn_input = attn_input.view(-1, attn_input.shape[-1])
        q_score = self.attn_score(attn_input)
        q_score = q_score.view(bsz, num_queries, -1)
        self_attn_rep = torch.multiply(q_score, v)
        summed_ha_rep = torch.sum(self_attn_rep, dim=-2)
        return summed_ha_rep


class LSTMHA(nn.Module):
    def __init__(self, lstm_input_size=FEATURE_DIM, lstm_h_dim=LSTM_H_DIM,
                 lstm_num_layers=LSTM_LAYER,
                 attn_w_dim=ATTN_W_DIM, dropout=LSTM_DROPOUT):
        super().__init__()

        self.lstm = nn.LSTM(input_size=lstm_input_size,
                            hidden_size=lstm_h_dim,
                            num_layers=lstm_num_layers,
                            batch_first=True,
                            bidirectional=False,
                            dropout=dropout)
        self.ha = StdAttn(input_size=lstm_h_dim,
                          att_dim=attn_w_dim, dropout=dropout)

    def forward(self, x):
        # x: (B, I, K, F)
        batch_size = x.shape[0]
        num_stock = x.shape[1]
        # (B * I, K, F)
        x = x.view(batch_size * num_stock, x.shape[-2], x.shape[-1])
        outputs, _ = self.lstm(x) # (B * I, K, H)
        # (B * I, H)
        out_rep = self.ha(outputs, outputs[:,-1,:], outputs)
        # (B, I, H)
        out_rep = out_rep.view(batch_size, num_stock, -1)
        return out_rep


class BasicCANN(nn.Module):
    def __init__(self, num_heads=CANN_HEADS, embed_dim=LSTM_H_DIM):
        super().__init__()
        self.attn = nn.MultiheadAttention(num_heads=num_heads,
                                          embed_dim=embed_dim,)
        self.score = nn.Sequential(
            nn.Linear(embed_dim, 1),
            nn.Sigmoid())

    def forward(self, x):
        x_permuted = x.permute(1, 0, 2) # (I, B, H)
        cross_attn_rep, attn_weights = self.attn(x_permuted, x_permuted, x_permuted)
        # Permute back: (B, I, H)
        cross_attn_rep = cross_attn_rep.permute(1, 0, 2)
        # (B, I, 1)
        scores = self.score(cross_attn_rep)
        return scores, attn_weights

class PortfolioGenerator(nn.Module):
    def __init__(self, G):
        super().__init__()
        self.G = G # Number of "winner" stocks to buy

    def forward(self, x):
        """
        MODIFIED FOR INDIA (Long-Only):
        We only create the long portfolio (b+).
        :param x: Winner scores (B, I, 1)
        :return: portfolio: (B, I)
        """
        # x: (B, I, 1) -> (B, I)
        x = x.squeeze(-1)

        # (B, I)
        sorted_scores, sorted_indices = torch.sort(x, dim=-1, descending=True)

        # winners
        winner_scores = sorted_scores[:, :self.G] # (B, G)

        # Get the original indices of the winners
        winner_indices = sorted_indices[:, :self.G] # (B, G)

        winner_proportion = F.softmax(winner_scores, dim=-1) # (B, G)

        b_plus = torch.zeros_like(x, device=x.device)

        b_plus.scatter_(dim=1, index=winner_indices, src=winner_proportion)

        #final portfolio b_t+
        return b_plus, sorted_indices

#PGNetwork
class PGNetwork(nn.Module):
    def __init__(self, G):
        super(PGNetwork, self).__init__()
        self.lstm_ha = LSTMHA()
        self.cann = BasicCANN()
        self.portfolio_gen = PortfolioGenerator(G=G) # Pass G

    def forward(self, x):
        if x.dim() == 3:
            #single inf
            x_b = x.unsqueeze(0)
        else:
            x_b = x

        # (B, I, H)
        ha_rep = self.lstm_ha(x_b)
        # (B, I, 1)
        cann_score, attn_w = self.cann(ha_rep)
        # (B, I)
        portfolios, sorted_indices = self.portfolio_gen(cann_score)

        return portfolios, sorted_indices
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import pandas as pd
import time
import gymnasium as gym
from collections import deque

#Hyper Parameters
MAX_EPISODE = 2000
MAX_STEP = 1000
GAMMA = 0.95
LR = 0.01


DATA_PATH = 'india_nifty100.csv'
NIFTY_TICKERS = list(pd.read_csv(DATA_PATH)['ticker'].unique())
NUM_STOCKS = len(NIFTY_TICKERS)
FEATURE_DIM = 3               # PR, TV, VOL
LOOK_BACK_K = 12              # 12-month history
HOLDING_PERIODS_T = 12        # 12-month investment (for Sharpe Ratio)
STOCK_POOL_G = 10             # Buy top 10 winners as per BWSL
TRAIN_TEST_SPLIT = '2020-01-01'

MODEL_PATH = 'modelsave/india_net_best.pkl'
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")


class IndiaStockEnv(gym.Env):
    """
    A Gymnasium environment for the AlphaStock model on Indian data.
    """
    def __init__(self, data_path, look_back_k, holding_periods_t, feature_dim,
                 stock_pool, train_test_split_date):

        self.data = pd.read_csv(data_path, parse_dates=['date'])
        self.look_back_k = look_back_k
        self.holding_periods_t = holding_periods_t
        self.feature_dim = feature_dim
        self.stock_pool = stock_pool
        self.num_stocks = len(self.stock_pool)

        self.feature_cols = ['PR', 'TV', 'VOL']

        self.train_data = self.data[self.data['date'] < pd.to_datetime(train_test_split_date)]
        self.test_data = self.data[self.data['date'] >= pd.to_datetime(train_test_split_date)]

        self.current_data = self.train_data
        self.unique_dates = sorted(self.current_data['date'].unique())

        self.episode_length = self.look_back_k + self.holding_periods_t

        self.start_indices = [
            i for i in range(len(self.unique_dates) - self.episode_length + 1)
        ]

        self.action_space = gym.spaces.Box(low=0, high=1, shape=(self.num_stocks,), dtype=np.float32)
        self.observation_space = gym.spaces.Box(
            low=-np.inf, high=np.inf,
            shape=(self.num_stocks, self.look_back_k, self.feature_dim),
            dtype=np.float32
        )

    def _get_state(self, date_index):
        """Get the state: (num_stocks, K, num_features)"""
        state = np.zeros(self.observation_space.shape, dtype=np.float32)
        history_dates = self.unique_dates[date_index - self.look_back_k : date_index]
        state_data = self.current_data[self.current_data['date'].isin(history_dates)]

        for i, ticker in enumerate(self.stock_pool):
            stock_data = state_data[state_data['ticker'] == ticker].sort_values('date')
            if len(stock_data) == self.look_back_k:
                state[i, :, :] = stock_data[self.feature_cols].values

        return state

    def reset(self):
        self.current_step = 0
        self.episode_start_index = np.random.choice(self.start_indices)
        self.current_date_index = self.episode_start_index + self.look_back_k
        self.period_returns = []

        return self._get_state(self.current_date_index)

    def step(self, action):
        current_date = self.unique_dates[self.current_date_index]
        returns_data = self.current_data[self.current_data['date'] == current_date]

        z_t = np.zeros(self.num_stocks, dtype=np.float32)
        for i, ticker in enumerate(self.stock_pool):
            val = returns_data[returns_data['ticker'] == ticker]['next_return'].values
            if len(val) > 0:
                z_t[i] = val[0]

        R_t = np.dot(action, z_t)
        self.period_returns.append(R_t)

        self.current_step += 1
        self.current_date_index += 1

        done = False
        reward = 0.0

        if self.current_step == self.holding_periods_t:
            done = True
            if len(self.period_returns) > 1 and np.std(self.period_returns) != 0:
                A_T = np.mean(self.period_returns)
                V_T = np.std(self.period_returns)
                reward = A_T / V_T
            else:
                reward = 0.0

        next_state = self._get_state(self.current_date_index) if not done else None

        return next_state, reward, done, {}


LSTM_H_DIM = 128
ATTN_W_DIM = 64
LSTM_DROPOUT = 0.2
LSTM_LAYER = 1
CANN_HEADS = 4

class StdAttn(nn.Module):
    def __init__(self, input_size, att_dim, dropout):
        super().__init__()
        self.attention_dim = att_dim
        self.query_transform = nn.Linear(input_size, att_dim)
        self.key_transform = nn.Linear(input_size, att_dim)
        self.attn_score = nn.Sequential(
            nn.Tanh(),
            nn.Dropout(p=dropout),
            nn.Linear(att_dim, 1),
            nn.Softmax(dim=-1)
        )

    def forward(self, q, k, v):
        embed_dim = q.shape[-1]
        bsz = q.shape[0]
        num_queries = q.shape[-2]
        if not k.shape[-1] == embed_dim or not v.shape[-1] == embed_dim:
            raise ValueError("Assume identical embedding dim for q, k, v")
        q = q.reshape(num_queries, bsz, embed_dim)
        trans_q = self.query_transform(q)
        trans_k = self.key_transform(k)
        attn_input = trans_k + trans_q
        attn_input = attn_input.view(-1, attn_input.shape[-1])
        q_score = self.attn_score(attn_input)
        q_score = q_score.view(bsz, num_queries, -1)
        self_attn_rep = torch.multiply(q_score, v)
        summed_ha_rep = torch.sum(self_attn_rep, dim=-2)
        return summed_ha_rep

class LSTMHA(nn.Module):
    def __init__(self, lstm_input_size=FEATURE_DIM, lstm_h_dim=LSTM_H_DIM,
                 lstm_num_layers=LSTM_LAYER,
                 attn_w_dim=ATTN_W_DIM, dropout=LSTM_DROPOUT):
        super().__init__()
        self.lstm = nn.LSTM(input_size=lstm_input_size,
                            hidden_size=lstm_h_dim,
                            num_layers=lstm_num_layers,
                            batch_first=True,
                            bidirectional=False,
                            dropout=dropout)
        self.ha = StdAttn(input_size=lstm_h_dim,
                          att_dim=attn_w_dim, dropout=dropout)

    def forward(self, x):
        batch_size = x.shape[0]
        num_stock = x.shape[1]
        x = x.view(batch_size * num_stock, x.shape[-2], x.shape[-1])
        outputs, _ = self.lstm(x)
        out_rep = self.ha(outputs, outputs[:,-1,:], outputs)
        out_rep = out_rep.view(batch_size, num_stock, -1)
        return out_rep

class BasicCANN(nn.Module):
    def __init__(self, num_heads=CANN_HEADS, embed_dim=LSTM_H_DIM):
        super().__init__()
        self.attn = nn.MultiheadAttention(num_heads=num_heads,
                                          embed_dim=embed_dim,)
        self.score = nn.Sequential(
            nn.Linear(embed_dim, 1),
            nn.Sigmoid())

    def forward(self, x):
        # x: (B, I, H)
        # PyTorch MHA expects (SeqLen, Batch, Embed)
        x_permuted = x.permute(1, 0, 2) # (I, B, H)
        cross_attn_rep, attn_weights = self.attn(x_permuted, x_permuted, x_permuted)
        # Permute back: (B, I, H)
        cross_attn_rep = cross_attn_rep.permute(1, 0, 2)
        scores = self.score(cross_attn_rep)
        return scores, attn_weights

class PortfolioGenerator(nn.Module):
    def __init__(self, G):
        super().__init__()
        self.G = G

    def forward(self, x):
        x = x.squeeze(-1)
        sorted_scores, sorted_indices = torch.sort(x, dim=-1, descending=True)
        winner_scores = sorted_scores[:, :self.G]
        winner_indices = sorted_indices[:, :self.G]
        winner_proportion = F.softmax(winner_scores, dim=-1)
        b_plus = torch.zeros_like(x, device=x.device)
        b_plus.scatter_(dim=1, index=winner_indices, src=winner_proportion)
        return b_plus, sorted_indices

class PGNetwork(nn.Module):
    def __init__(self, G):
        super(PGNetwork, self).__init__()
        self.lstm_ha = LSTMHA()
        self.cann = BasicCANN()
        self.portfolio_gen = PortfolioGenerator(G=G)

    def forward(self, x):
        if x.dim() == 3:
            x_b = x.unsqueeze(0)
        else:
            x_b = x
        ha_rep = self.lstm_ha(x_b)
        cann_score, attn_w = self.cann(ha_rep)
        portfolios, sorted_indices = self.portfolio_gen(cann_score)
        return portfolios, sorted_indices


class PG(object):
    """
    Policy Gradient agent that optimizes for the *episode* reward.
    The episode reward is the Sharpe Ratio (H_T).
    """
    def __init__(self, num_stocks, G):
        self.ep_obs, self.ep_as = [], []
        self.network = PGNetwork(G=G).to(device)
        self.optimizer = torch.optim.Adam(self.network.parameters(), lr=LR)

    def choose_action(self, observation):
        observation = torch.FloatTensor(observation).to(device)
        network_output, _ = self.network.forward(observation)
        action_prob = network_output.detach().cpu().numpy()[0]
        return action_prob

    def store_transition(self, s, a):
        self.ep_obs.append(s)
        self.ep_as.append(a)

    def update_parameters(self, episode_reward):
        H_pi = episode_reward
        H_0 = 0.0
        reward_weight = H_pi - H_0

        if reward_weight == 0 or len(self.ep_obs) == 0:
            self.ep_obs, self.ep_as = [], []
            return 0.0

        obs_tensor = torch.FloatTensor(np.array(self.ep_obs)).to(device)
        portfolios_pred, _ = self.network.forward(obs_tensor)

        log_G = torch.log(portfolios_pred[portfolios_pred > 0] + 1e-9)

        loss = - (reward_weight * torch.sum(log_G))

        self.optimizer.zero_grad()
        loss.backward()
        self.optimizer.step()

        self.ep_obs, self.ep_as = [], []
        return loss.item()


def train():
    env = IndiaStockEnv(
        data_path=DATA_PATH,
        look_back_k=LOOK_BACK_K,
        holding_periods_t=HOLDING_PERIODS_T,
        feature_dim=FEATURE_DIM,
        stock_pool=NIFTY_TICKERS,
        train_test_split_date=TRAIN_TEST_SPLIT
    )

    agent = PG(num_stocks=NUM_STOCKS, G=STOCK_POOL_G)

    print(f"Starting training on {NUM_STOCKS} stocks...")

    max_train_reward = -np.inf

    for episode in range(1, MAX_EPISODE + 1):
        state = env.reset()
        start_time = time.time()

        for step in range(HOLDING_PERIODS_T):
            action = agent.choose_action(state)
            next_state, reward, done, _ = env.step(action)

            agent.store_transition(state, action)
            state = next_state

            if done:
                loss = agent.update_parameters(episode_reward=reward)

                if episode % 20 == 0:
                    print(f"Episode: {episode}, Reward (Sharpe): {reward:.4f}, "
                          f"Loss: {loss:.4f}, Time: {time.time() - start_time:.2f}s")

                if reward > max_train_reward:
                    max_train_reward = reward
                    torch.save(agent.network.state_dict(), MODEL_PATH)
                    print(f"*** New best model saved with Sharpe: {reward:.4f} at episode {episode} ***")

                break


if __name__ == '__main__':
    train()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def calculate_metrics(returns_series, periods_per_year=12, tc_per_trade=0.001):
    """
    Calculates all metrics from the AlphaStock paper
    given a pandas Series of periodic returns (R_t).

    NOW ALSO RETURNS the drawdown_series for plotting.
    """
    # Apply transaction cost
    returns_series = returns_series - tc_per_trade

    if returns_series.empty or len(returns_series) < 2:
        return {
            "APR": 0, "AVOL": 0, "ASR": 0,
            "MDD": 0, "CR": 0, "DDR": 0, "Cumulative Wealth": 1
        }, None, None

    # 1. Annualized Percentage Rate (APR)
    apr = returns_series.mean() * periods_per_year

    # 2. Annualized Volatility (AVOL)
    avol = returns_series.std() * np.sqrt(periods_per_year)

    # 3. Annualized Sharpe Ratio (ASR)
    asr = apr / avol if avol != 0 else 0

    # 4. Cumulative Wealth
    cumulative_wealth = (1 + returns_series).cumprod()

    # 5. Maximum DrawDown (MDD)
    running_max = cumulative_wealth.cummax()
    drawdown_series = (cumulative_wealth - running_max) / running_max # <-- This is the time-series
    mdd = drawdown_series.min() # <-- This is the single scalar value

    # 6. Calmar Ratio (CR)
    cr = apr / abs(mdd) if mdd != 0 else 0

    # 7. Downside Deviation Ratio (DDR)
    downside_returns = returns_series[returns_series < 0]
    if not downside_returns.empty:
        expected_squared_downside = (downside_returns**2).mean()
        annualized_downside_dev = np.sqrt(expected_squared_downside) * np.sqrt(periods_per_year)
        ddr = apr / annualized_downside_dev if annualized_downside_dev != 0 else 0
    else:
        ddr = 0

    metrics = {
        "APR": apr,
        "AVOL": avol,
        "ASR": asr,
        "MDD": mdd,
        "CR": cr,
        "DDR": ddr,
        "Cumulative Wealth": cumulative_wealth.iloc[-1]
    }

    # Return the new drawdown_series as well
    return metrics, cumulative_wealth, drawdown_series

def plot_cumulative_wealth(cumulative_wealth_series, title='Cumulative Wealth Over Time'):
    """
    Plots the cumulative wealth over time, similar to Figure 2.
    """
    if cumulative_wealth_series is None:
        print("Cannot plot: No cumulative wealth data.")
        return

    plt.figure(figsize=(12, 6))
    cumulative_wealth_series.plot()
    plt.title(title)
    plt.xlabel('Time')
    plt.ylabel('Cumulative Wealth (Log Scale)')
    plt.yscale('log') # Use log scale
    plt.grid(True, which="both", ls="--", linewidth=0.5)
    plt.show()

def plot_drawdown(drawdown_series, title='Portfolio Drawdown Over Time'):
    """
    Plots the portfolio's drawdown over time.
    """
    if drawdown_series is None:
        print("Cannot plot: No drawdown data.")
        return

    plt.figure(figsize=(12, 6))
    # Plot as a filled area chart
    plt.fill_between(drawdown_series.index, drawdown_series, 0, color='red', alpha=0.3)
    drawdown_series.plot(color='red')
    plt.title(title)
    plt.xlabel('Time')
    plt.ylabel('Drawdown')
    plt.grid(True, which="both", ls="--", linewidth=0.5)
    plt.show()

def plot_scalar_metrics(metrics, title='Final Performance Metrics'):
    """
    Plots the scalar metrics (APR, ASR, etc.) on a bar chart.
    """
    # We multiply percentages by 100 to make the bar chart readable
    plot_data = {
        'APR (%)': metrics.get('APR', 0) * 100,
        'AVOL (%)': metrics.get('AVOL', 0) * 100,
        'ASR': metrics.get('ASR', 0),
        'MDD (%)': metrics.get('MDD', 0) * 100,
        'CR': metrics.get('CR', 0),
        'DDR': metrics.get('DDR', 0)
    }

    names = list(plot_data.keys())
    values = list(plot_data.values())

    plt.figure(figsize=(10, 6))
    bars = plt.barh(names, values)

    plt.xlabel('Value (Ratios or %)')
    plt.title(title)
    plt.grid(True, which="both", ls="--", linewidth=0.5, axis='x')

    # Add data labels to the bars
    for bar in bars:
        width = bar.get_width()
        plt.text(width + 0.05 * np.sign(width), # Position label
                 bar.get_y() + bar.get_height()/2,
                 f'{width:.2f}',
                 va='center')

    plt.gca().invert_yaxis() # Puts APR at the top
    plt.show()
```

```python


class PGtest(object):
    """
    A simple class to load and run the trained model for inference.
    """
    def __init__(self, model_path, G):
        # Assumes PGNetwork is defined (in your training cell)
        self.network = PGNetwork(G=G).to(device)
        self.network.load_state_dict(torch.load(model_path, map_location=device))
        self.network.eval()
        print(f"Model loaded from {model_path}")

    @torch.no_grad()
    def choose_action(self, observation):
        observation = torch.FloatTensor(observation).to(device).unsqueeze(0)
        network_output, _ = self.network.forward(observation)
        return network_output.detach().cpu().numpy()[0]


def test():

    # Assumes IndiaStockEnv and params are defined (in your training cell)
    test_env = IndiaStockEnv(
        data_path=DATA_PATH,
        look_back_k=LOOK_BACK_K,
        holding_periods_t=HOLDING_PERIODS_T,
        feature_dim=FEATURE_DIM,
        stock_pool=NIFTY_TICKERS,
        train_test_split_date=TRAIN_TEST_SPLIT
    )

    test_env.current_data = test_env.test_data
    test_env.unique_dates = sorted(test_env.current_data['date'].unique())

    agent = PGtest(MODEL_PATH, STOCK_POOL_G)

    all_returns = []
    start_index = LOOK_BACK_K

    if len(test_env.unique_dates) <= start_index:
        print("Error: Test data is too short for the look-back period.")
        return

    print(f"Test data from {test_env.unique_dates[start_index].date()} to {test_env.unique_dates[-1].date()}")

    for t in range(start_index, len(test_env.unique_dates) - 1):
        state = test_env._get_state(t)
        action = agent.choose_action(state)

        current_date = test_env.unique_dates[t]
        returns_data = test_env.current_data[test_env.current_data['date'] == current_date]

        z_t = np.zeros(NUM_STOCKS, dtype=np.float32)
        for i, ticker in enumerate(NIFTY_TICKERS):
            val = returns_data[returns_data['ticker'] == ticker]['next_return'].values
            if len(val) > 0:
                z_t[i] = val[0]

        R_t = np.dot(action, z_t)
        all_returns.append(R_t)

    print(f"Backtest complete. Calculated {len(all_returns)} monthly returns.")

    if not all_returns:
        print("No returns were calculated. Cannot generate metrics.")
        return

    returns_series = pd.Series(
        all_returns,
        index=pd.to_datetime(test_env.unique_dates[start_index : start_index + len(all_returns)])
    )



    # 1. Calculate metrics (now gets drawdown_series)
    metrics, cumulative_wealth_series, drawdown_series = calculate_metrics(returns_series, tc_per_trade=0.001)


    print("\n Backtest Results")
    print(f"Cumulative Wealth:        {metrics['Cumulative Wealth']:.4f}")
    print(f"APR (Annualized Return):  {metrics['APR']:.2%}")
    print(f"AVOL (Annualized Vol):    {metrics['AVOL']:.2%}")
    print(f"ASR (Annualized Sharpe):  {metrics['ASR']:.4f}")
    print(f"MDD (Max Drawdown):       {metrics['MDD']:.2%}")
    print(f"CR (Calmar Ratio):        {metrics['CR']:.4f}")
    print(f"DDR (Downside Deviation): {metrics['DDR']:.4f}")


    print("\nResults")
    plot_cumulative_wealth(cumulative_wealth_series, title="AlphaStock (India) Cumulative Wealth")
    plot_drawdown(drawdown_series, title="AlphaStock (India) Drawdown")
    plot_scalar_metrics(metrics, title="AlphaStock (India) Final Metrics")


test()
```
